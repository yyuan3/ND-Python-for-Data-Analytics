### MATLAB-Style vs. Object Oriented Interface
One of the reason that Matplotlib can be confusing to beginners (or anyone who isn't constantly using it) is that it presents to different interfaces. Your textbook opts for relying on the MATLAB style interface, but we will not do so.

We will stick with the object-oriented approach, which is more consistent with the language as a whole and gives you more control over your plots.

### The Anatomy of a Plot
Before we go any farther, let's define some Matplotlib terms:
* Every plot/chart is made of one **`figure`** object and 1 or more **`axes`** objects
* The **`figure`** object is more or less just a container
* Each **`axes`** object is a graphical representation of data that exists inside the figure.
    * Most often, a *`figure`* will contain only one *`axes`* but that is not always the case as we will learn in later.

## Simple Line Plots

```python
# Load and set the index column to be the player's jersey number.
nd_football_roster = pd.read_csv('http://www3.nd.edu/~jng2/nd-football-roster-2019.csv', index_col='Number')
nd_football_roster.head()
```

In order to create a plot, we have to have the two essential components mentioned above - a **`figure`** object and a **`axes`** object.

### Plotting Your First Data
Let's plot the heights of the players.

```python
# Get the objects
figure, axes = plt.subplots()

# Use the `plot()` method to "write" data on your axes.
# The first argument is used as the x-axis
# The second is used for the y-axis
axes.plot(nd_football_roster.index, nd_football_roster['Height'])
```

You can display multiple datasets on a single **`axes`** object. Let's divide the *`nd_football_roster['Height']`* series into two parts and display them as separate lines.

### GroupBy 'day_of_week' and Plotting

We want to see if some days of the week have more crimes than other days.

1. Compute the number of crimes per day of the week
2. Plot number of crimes vs day of week
3. What do you interpret from the visualization? 
    

```python
crime = pd.read_csv("http://www3.nd.edu/~jng2/chicago-recent-crime.csv", parse_dates=['DATE OF OCCURRENCE'])
crime.head()
```

```python
crime['day_of_week'] = crime['DATE OF OCCURRENCE'].dt.day_name()
crime.head()
```

```python
crimes_by_DoW = crime.groupby(['day_of_week'])[['DATE OF OCCURRENCE']].count().rename(
    columns={'DATE OF OCCURRENCE': 'num_crimes'})
crimes_by_DoW.reset_index(inplace=True)
crimes_by_DoW.head()
```

```python
figure, axes = plt.subplots()
axes.plot(crimes_by_DoW['day_of_week'], crimes_by_DoW['num_crimes'])
```

Notice that the x axis is not in the order you would expect. That's because 'day_of_week' 
was sorted alphabetically by default. 

Let's sort 'day_of_week' in the order we want by calling groupby on dt.dayofweek

```python
crimes_by_DoW_2 = crime.groupby( 
    crime['DATE OF OCCURRENCE'].dt.dayofweek )[['DATE OF OCCURRENCE']].count().rename(
    columns={'DATE OF OCCURRENCE': 'num_crimes'})
crimes_by_DoW_2.reset_index(inplace=True)
crimes_by_DoW_2.head()
```

```python
# The 'DATE OF OCCURRENCE' column is misleadingly labeled. Rename it.

crimes_by_DoW_2.rename(columns={'DATE OF OCCURRENCE' : 'dayofweek'}, inplace=True)
```

```python
figure, axes = plt.subplots()
axes.plot(crimes_by_DoW_2['dayofweek'] , crimes_by_DoW_2['num_crimes'])
```

```python
# The code above is kind of cluttered. Might be good to break it up.
dow = crime['DATE OF OCCURRENCE'].dt.dayofweek
crime_groupedby_dayofweek = crime.groupby( dow )
crimes_by_DoW_3 = crime_groupedby_dayofweek[['DATE OF OCCURRENCE']].count().rename(columns={'DATE OF OCCURRENCE': 'num_crimes'})
crimes_by_DoW_3.reset_index(inplace=True)
crimes_by_DoW_3
```

### Adjusting Line Properties
There are a number of line properties that you can adjust via keyword parameters to the *`plot()`*. 

Let's explore a few of them here: **`alpha`, `color`, `linestyle`, `linewidth`, `marker` and `markersize`.**

```python
figure, axes = plt.subplots()

# The alpha parameter determines how transparent a line will be.
axes.plot(np.arange(0,50), nd_football_roster['Height'][:50], alpha=.25)
axes.plot(np.arange(0,50), nd_football_roster['Height'][50:100], alpha=.75)
```

Matplotlib allows you to specify color choices in a very wide variety of ways. In the following example, I'll divide up our players into groups of ten to show some of the various ways to specify line color.

```python
figure, axes = plt.subplots()

# the color parameter controls the color of your line.

# You can specify any HTML color. 
# https://www.w3schools.com/colors/colors_names.asp
axes.plot(np.arange(0,10), nd_football_roster['Height'][:10], 
          color='CadetBlue')
axes.plot(np.arange(0,10), nd_football_roster['Height'][10:20],
          color='OliveDrab')
```

```python
figure, axes = plt.subplots()

# Floats between 0 and 1 specify grayscale values
# But notice they have to be passed as strings
axes.plot(np.arange(0,10), nd_football_roster['Height'][20:30], 
          color='.1')
axes.plot(np.arange(0,10), nd_football_roster['Height'][30:40], 
          color='.8')
```

```python
figure, axes = plt.subplots()

# Hex Codes are acceptable (#RRGGBB)
# Google for: hex color picker
axes.plot(np.arange(0,10), nd_football_roster['Height'][40:50], 
          color='#f4429e')
axes.plot(np.arange(0,10), nd_football_roster['Height'][50:60], 
          color='#af7313')
```

```python
figure, axes = plt.subplots()

# You can even specify RGB tuples
# Each component of the tuple is a float between 0 and 1
axes.plot(np.arange(0,10), nd_football_roster['Height'][60:70], 
          color=(1.0, 0, 0)) # Total Red
axes.plot(np.arange(0,10), nd_football_roster['Height'][70:80], 
          color=(1.0, 0, 1.0)) # Pure Purple
axes.plot(np.arange(0,10), nd_football_roster['Height'][80:90], 
          color=(0.0, 1.0, 0.0)) # Gorgeous Green
```

#### Line Style
The *`linestyle`* parameter can take any of the options in the following table. You can specify either the or the symbol of the option:

| Name | Symbol|
|--------|-------|
| solid  | -     |
| dashed | --    |
| dashdot  | -.     |
| dotted  | :     |
|--------|-------|

```python
figure, axes = plt.subplots()

# Specifying with name
axes.plot(np.arange(0,10), nd_football_roster['Height'][:10], 
          color='CadetBlue', linestyle='dashed')
axes.plot(np.arange(0,10), nd_football_roster['Height'][10:20], 
          linestyle='dotted')

# Specifying with symbol
axes.plot(np.arange(0,10), nd_football_roster['Height'][20:30], 
          linestyle='-.')
axes.plot(np.arange(0,10), nd_football_roster['Height'][30:40], 
          linestyle='-', alpha=.6)
```

#### Line Width

```python
# Control Line Width with the `linewidth` parameter
figure, axes = plt.subplots()

axes.plot(np.arange(0,10), nd_football_roster['Height'][:10], 
          color='CadetBlue', linestyle='dashed', linewidth=4.5)

axes.plot(np.arange(0,10), nd_football_roster['Height'][10:20], 
          linestyle='dotted', linewidth=8)

axes.plot(np.arange(0,10), nd_football_roster['Height'][20:30], 
          linestyle='-.', linewidth=2.8)
```

#### Markers
If you so desire, you can add a "mark" at each data point in your plots through a combination of the *`marker`* and *`markersize`* parameters. 

All of the available marker styles are <a href='https://matplotlib.org/api/markers_api.html#module-matplotlib.markers' target='_blank'>detailed in Matplotlib's online documentation.</a>

Here are some examples of how to use these two parameters.

```python
figure, axes = plt.subplots()

# Use diamond markers
axes.plot(np.arange(0,10), nd_football_roster['Height'][:10], 
          marker='D', markersize=5)

# You can combine linestyle and markers
# Here we will use a star marker
axes.plot(np.arange(0,10), nd_football_roster['Height'][10:20], 
          linestyle='dotted', marker='*', markersize=20)

# Circle Markers
axes.plot(np.arange(0,10), nd_football_roster['Height'][20:30],
          marker='o', markersize=5)
```

### Adjusting Axes Limits
If we go back to our original plot, we can see that Matplotlib automatically sets the limits of the x-axis and y-axis values based on the data that we give it.

```python
figure, axes = plt.subplots()
axes.plot(np.arange(0, 111), nd_football_roster['Height'])
```

You do not have to accept these default values however. Using the *`set_xlim`* and *`set_ylim`* methods of the **`axes`** object, we can visually focus on a subset of the data.

```python
figure, axes = plt.subplots()

# Let's display the sorted weights of the players on the team.
axes.plot(np.arange(0, 111), 
          nd_football_roster['Weight'].sort_values())
```

```python
figure, axes = plt.subplots()

# And then use the `set_ylim` method to focus in on the players 
# who weigh between 250 and 300 lbs and whose x-axis values
# are between 80 and 90
axes.plot(np.arange(0, 111), 
          nd_football_roster['Weight'].sort_values(), 
          marker='o')

axes.set_ylim(250, 300)
axes.set_xlim(80, 90)
```

### Adding Plot Metadata (Labels)
Now that we've covered the basics of plotting lines, let's turn our attention to adding plot metadata.

#### Plot Titles

```python
# Add plot title with axes.set_title() method.
figure, axes = plt.subplots()
axes.plot(nd_football_roster.index, nd_football_roster['Height'])
axes.set_title(label='2019 ND Football Player Heights')
```

```python
# You can adjust the horizontal alignment 
# of the title with the `loc` parameter.
# The value must be one of the following: left, center, right
figure, axes = plt.subplots()
axes.plot(nd_football_roster.index, nd_football_roster['Height'])

# Notice that if you specify location, 
# you can have multiple titles.
axes.set_title(label='2019 ND Football', loc='left')
axes.set_title(label='Player Heights', loc='right')
```

```python
figure, axes = plt.subplots()
axes.plot(nd_football_roster.index, nd_football_roster['Height'])

# Here we will adjust the color 
axes.set_title(label='2019 ND Football', loc='left', color='gold')
axes.set_title(label='Player Heights', loc='right', 
               color='darkblue')
```

#### Axis Labels
Our plot is starting to look pretty cool, but we don't really know what the units of measurement are. We can fix that with the *`set_ylabal()`* and *`set_xlabel`* methods.

```python
figure, axes = plt.subplots()
axes.plot(nd_football_roster.index, nd_football_roster['Height'])

axes.set_title(label='2019 ND Football', loc='left')
axes.set_title(label='Player Heights', loc='right')

axes.set_ylabel('Height(in)')
axes.set_xlabel('Jersey #')
```

With that additional information, our plot can be understood by other users. You can also control the color and opacity of these labels just as you can with the plot titles:

```python
figure, axes = plt.subplots()
axes.plot(nd_football_roster.index, nd_football_roster['Height'])

axes.set_title(label='2019 ND Football', loc='left')
axes.set_title(label='Player Heights', loc='right')

axes.set_ylabel('Height(in)', color='red')
axes.set_xlabel('Jersey #')
```

#### Plot Legends
In order to complete our plot, we need one more thing: a legend.
Like most things in Matplotlib, adding a basic one is simple - but with many customization options.

```python
figure, axes = plt.subplots()

# We will divide our team into two squads for
# illustrative purposes.
axes.plot(np.arange(0,50), nd_football_roster['Height'][:50], 
          linestyle='--')
axes.plot(np.arange(0,50), nd_football_roster['Height'][50:100])

axes.set_title(label='2019 ND Football', loc='left')
axes.set_title(label='Player Heights', loc='right')

axes.set_ylabel('Height(in)')
axes.set_xlabel('Jersey #')

# Initialize the plot legend 
axes.legend()
```

Let's take note of a few things that Matplotlib does for the default legend:
1. It generates a legend item from each plotted line.
2. It matches the style of the line with the style in the legend.
3. It derives the name of each legend from the data being plotted if
possible.

As you can see, having both lines labels as 'Height' isn't the most ideal. Let's manually set a label for each of our lines so that the **`legend()`** method can pick them up.

```python
figure, axes = plt.subplots()

# We set the labels for each line.
axes.plot(np.arange(0,50), nd_football_roster['Height'][:50], 
          linestyle='--', label='Squad A')
axes.plot(np.arange(0,50), nd_football_roster['Height'][50:100], 
          label='Squad B')

axes.set_title(label='2019 ND Football', loc='left')
axes.set_title(label='Player Heights', loc='right')

axes.set_ylabel('Height(in)')
axes.set_xlabel('Jersey #')

# Initialize the plot legend 
axes.legend()
```

There are a number of other arguments/parameters that you can specify to further customize plot legends. Let's go over a few of them:

```python
figure, axes = plt.subplots()

# We set the labels for each line.
axes.plot(np.arange(0,50), nd_football_roster['Height'][:50], 
          linestyle='--', label='Squad A')
axes.plot(np.arange(0,50), nd_football_roster['Height'][50:100], 
          label='Squad B')

axes.set_title(label='2019 ND Football', loc='left')
axes.set_title(label='Player Heights', loc='right')

axes.set_ylabel('Height(in)')
axes.set_xlabel('Jersey #')

# Use the `loc` argument to control where the
# legend appears in your plot. Valid options:
# upper left, upper center, upper right
# center left, center, center right
# lower left, lower center, lower right
axes.legend(loc='lower center')
```

You can set a title for the legend itself:

```python
figure, axes = plt.subplots()
axes.plot(np.arange(0,50), nd_football_roster['Height'][:50], 
          linestyle='--', label='Squad A')
axes.plot(np.arange(0,50), nd_football_roster['Height'][50:100], 
          label='Squad B')

axes.set_title(label='2019 ND Football', loc='left')
axes.set_title(label='Player Heights', loc='right')

axes.set_ylabel('Height(in)')
axes.set_xlabel('Jersey #')

axes.legend(loc='lower center', 
            title='I Am Legend')
```

You can add edges and background colors to a legend:

```python
figure, axes = plt.subplots()
axes.plot(np.arange(0,50), nd_football_roster['Height'][:50], 
          linestyle='--', label='Squad A')
axes.plot(np.arange(0,50), nd_football_roster['Height'][50:100], 
          label='Squad B')

axes.set_title(label='2019 ND Football', loc='left')
axes.set_title(label='Player Heights', loc='right')

axes.set_ylabel('Height(in)')
axes.set_xlabel('Jersey #')

# You can specify background and edge colors for the 
# legend if you also specify `frameon=True`
axes.legend(title='I Am Legend', frameon=True, 
            edgecolor='black', facecolor='gold')
```

You can adjust the number of columns in a legend. This is handy if you want the legend to take up less vertical space.

```python
figure, axes = plt.subplots()
axes.plot(np.arange(0,50), nd_football_roster['Height'][:50], 
          linestyle='--', label='Squad A')
axes.plot(np.arange(0,50), nd_football_roster['Height'][50:100], 
          label='Squad B')

axes.set_title(label='2019 ND Football', loc='left')
axes.set_title(label='Player Heights', loc='right')

axes.set_ylabel('Height(in)')
axes.set_xlabel('Jersey #')

# Use the `ncol` argument to specify the
# number of columns.
axes.legend(ncol=2, loc='lower right')
```

And finally, you can adjust the size of the font used in the legend:

```python
figure, axes = plt.subplots()
axes.plot(np.arange(0,50), nd_football_roster['Height'][:50], 
          linestyle='--', label='Squad A')
axes.plot(np.arange(0,50), nd_football_roster['Height'][50:100], 
          label='Squad B')

axes.set_title(label='2019 ND Football', loc='left')
axes.set_title(label='Player Heights', loc='right')

axes.set_ylabel('Height(in)')
axes.set_xlabel('Jersey #')

# Use `fontsize` to adjust the legend font size.
axes.legend(ncol=2, loc='lower right', fontsize=13)
```

## Generating subplots

Plot number of crimes by day of week separately for sidewalk, alley, street, apartment

```python
crimes_by_DoW_4 = crime.groupby( [crime['DATE OF OCCURRENCE'].dt.dayofweek, 
                                  'LOCATION DESCRIPTION']  )[['DATE OF OCCURRENCE']].count().rename(
    columns={'DATE OF OCCURRENCE': 'num_crimes'})
crimes_by_DoW_4.reset_index(inplace=True)
crimes_by_DoW_4.rename(columns={'DATE OF OCCURRENCE': 'dayofweek'}, inplace=True)
```

```python
crimes_by_DoW_4.head()
```

```python
crimes_sidewalk = crimes_by_DoW_4[ crimes_by_DoW_4['LOCATION DESCRIPTION'] == 'SIDEWALK' ]
crimes_alley = crimes_by_DoW_4[ crimes_by_DoW_4['LOCATION DESCRIPTION'] == 'ALLEY' ]
crimes_street = crimes_by_DoW_4[ crimes_by_DoW_4['LOCATION DESCRIPTION'] == 'STREET' ]
crimes_apartment = crimes_by_DoW_4[ crimes_by_DoW_4['LOCATION DESCRIPTION'] == 'APARTMENT' ]
```

```python
figure, axes = plt.subplots(1,4)
```

Notice that the *`axes`* variable now points to a Numpy array filled with **`AxesSubplot`** objects. If we attempt to plot something on the *`axes`* variable itself it will fail because a Numpy array can't plot anything!

```python
figure, axes = plt.subplots(1,4)
axes.plot(crimes_sidewalk['dayofweek'], crimes_sidewalk['num_crimes'], marker='o')
```

Instead we have to address the specific AxesSubplot objects that we want to plot with:

```python
figure, axes = plt.subplots(1,4)
axes[0].plot(crimes_sidewalk['dayofweek'], crimes_sidewalk['num_crimes'], marker='o')
axes[0].set_title('Sidewalk')
axes[1].plot(crimes_alley['dayofweek'], crimes_alley['num_crimes'], marker='o')
axes[1].set_title('Alley')
axes[2].plot(crimes_street['dayofweek'], crimes_street['num_crimes'], marker='o')
axes[2].set_title('Street')
axes[3].plot(crimes_apartment['dayofweek'], crimes_apartment['num_crimes'], marker='o')
axes[3].set_title('Apartment')
plt.tight_layout() # what's this line for? comment it out to find out
```

A potentially confusing case arises when you create a grid with multiple rows as this results in a 2 dimensional Numpy array being returned from the plt.subplots() method.

When you have this condition, you have to specify both the row and column indices of the AxesSubplot objects you want to plot with.

```python
figure, axes = plt.subplots(2,2)
axes[0,0].plot(crimes_sidewalk['dayofweek'], crimes_sidewalk['num_crimes'], marker='o')
axes[0,0].set_title('Sidewalk')
axes[0,1].plot(crimes_alley['dayofweek'], crimes_alley['num_crimes'], marker='o')
axes[0,1].set_title('Alley')
axes[1,0].plot(crimes_street['dayofweek'], crimes_street['num_crimes'], marker='o')
axes[1,0].set_title('Street')
axes[1,1].plot(crimes_apartment['dayofweek'], crimes_apartment['num_crimes'], marker='o')
axes[1,1].set_title('Apartment')
plt.tight_layout()
```

### Sharing X/Y Axis between Subplots

```python
figure, axes = plt.subplots(1, 4, sharey=True)
axes[0].plot(crimes_sidewalk['dayofweek'], crimes_sidewalk['num_crimes'], marker='o')
axes[0].set_title('Sidewalk')
axes[1].plot(crimes_alley['dayofweek'], crimes_alley['num_crimes'], marker='o')
axes[1].set_title('Alley')
axes[2].plot(crimes_street['dayofweek'], crimes_street['num_crimes'], marker='o')
axes[2].set_title('Street')
axes[3].plot(crimes_apartment['dayofweek'], crimes_apartment['num_crimes'], marker='o')
axes[3].set_title('Apartment')
plt.tight_layout()
```

### Adding a Title to a Plot with Multiple Subplots

```python
# Specify suptitle. 

figure, axes = plt.subplots(1, 4, sharey=True)
axes[0].plot(crimes_sidewalk['dayofweek'], crimes_sidewalk['num_crimes'], marker='o')
axes[0].set_title('Sidewalk')
axes[1].plot(crimes_alley['dayofweek'], crimes_alley['num_crimes'], marker='o')
axes[1].set_title('Alley')
axes[2].plot(crimes_street['dayofweek'], crimes_street['num_crimes'], marker='o')
axes[2].set_title('Street')
axes[3].plot(crimes_apartment['dayofweek'], crimes_apartment['num_crimes'], marker='o')
axes[3].set_title('Apartment')
plt.suptitle("Num Crimes by Day of Week at various locations")
plt.tight_layout()

# Notice that the positioning of the title is not good, it eats into the subplot area.
```

```python
# Easiest way to correct the positioning of suptitle is by setting constrained_layout = True in the subplots call.

figure, axes = plt.subplots(1, 4, sharey=True, constrained_layout = True)
axes[0].plot(crimes_sidewalk['dayofweek'], crimes_sidewalk['num_crimes'], marker='o')
axes[0].set_title('Sidewalk')
axes[1].plot(crimes_alley['dayofweek'], crimes_alley['num_crimes'], marker='o')
axes[1].set_title('Alley')
axes[2].plot(crimes_street['dayofweek'], crimes_street['num_crimes'], marker='o')
axes[2].set_title('Street')
axes[3].plot(crimes_apartment['dayofweek'], crimes_apartment['num_crimes'], marker='o')
axes[3].set_title('Apartment')
plt.suptitle("Number of crimes by day of week at various locations")

# Note that tight_layout() is incompatible with constrained_layout, so we've removed it.
```
